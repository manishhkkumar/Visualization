```python
#Visualization - Matplotlib and Seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('cars (3).csv', skipinitialspace=True)
```

```python
# Create scatter plots for the following data, make sure all the plots appear in the same
# plane with x-labels and y-labels evenly spaced.
# a. Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
# hp-time-to-60) with respect to each brand.
```

```python
fig = plt.subplots(2,2, figsize=(20,10))

plt.subplot(2,2,1)
sns.scatterplot(x=data1['mpg'],y=data1['hp'],data=data1, hue = data1['brand'])
plt.title('mpg vs hp')
plt.xlabel('mpg')
plt.ylabel('hp')

plt.subplot(2,2,2)
sns.scatterplot(x=data1['mpg'],y=data1['weightlbs'],data=data1, hue = data1['brand'])
plt.title('mpg vs weightlbs')
plt.xlabel('mpg')
plt.ylabel('weightlbs')

plt.subplot(2,2,3)
sns.scatterplot(x=data1['hp'],y=data1['mpg'],data=data1, hue = data1['brand'])
plt.title('hp vs mpg')
plt.xlabel('hp')
plt.ylabel('mpg')

plt.subplot(2,2,4)
sns.scatterplot(x=data1['hp'],y=data1['time-to-60'],data=data1, hue = data1['brand'])
plt.title('hp vs time-to-60')
plt.xlabel('hp')
plt.ylabel('time-to-60')

plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.show()
```

```python
# Create bar plots for the following, make sure all the plots appear in the same plane with
# x-labels and y-labels evenly spaced.
# Create a bar plot that shows the visual representation of hp, mpg, weight,
# time-to-60 with respect to the number of cylinders for each of the brands.
```

```python
fig = plt.subplots(2,2, figsize=(20,10))

plt.subplot(2,2,1)
sns.barplot(x=data1['cylinders'],y=data1['hp'],data=data1, hue = data1['brand'])
plt.title('cylinders vs hp')
plt.xlabel('cylinders')
plt.ylabel('hp')

plt.subplot(2,2,2)
sns.barplot(x=data1['cylinders'],y=data1['weightlbs'],data=data1, hue = data1['brand'])
plt.title('cylinders vs weightlbs')
plt.xlabel('cylinders')
plt.ylabel('weightlbs')

plt.subplot(2,2,3)
sns.barplot(x=data1['cylinders'],y=data1['mpg'],data=data1, hue = data1['brand'])
plt.title('cylinders vs mpg')
plt.xlabel('cylinders')
plt.ylabel('mpg')

plt.subplot(2,2,4)
sns.barplot(x=data1['cylinders'],y=data1['time-to-60'],data=data1, hue = data1['brand'])
plt.title('cylinders vs time-to-60')
plt.xlabel('cylinders')
plt.ylabel('time-to-60')

plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.show()
```

```python
#  Create box plots for the following:
#  a. Create box plots for the columns in the step1 with respect to the number of
#  cylinders and brand respectively.
```

```python
fig = plt.subplots(2,2, figsize=(20,10))

plt.subplot(2,2,1)
sns.boxplot(x=data1['cylinders'],y=data1['hp'],data=data1, hue = data1['brand'])
plt.title('cylinders vs hp')
plt.xlabel('cylinders')
plt.ylabel('hp')

plt.subplot(2,2,2)
sns.boxplot(x=data1['cylinders'],y=data1['weightlbs'],data=data1, hue = data1['brand'])
plt.title('cylinders vs weightlbs')
plt.xlabel('cylinders')
plt.ylabel('weightlbs')

plt.subplot(2,2,3)
sns.boxplot(x=data1['cylinders'],y=data1['mpg'],data=data1, hue = data1['brand'])
plt.title('cylinders vs mpg')
plt.xlabel('cylinders')
plt.ylabel('mpg')

plt.subplot(2,2,4)
sns.boxplot(x=data1['cylinders'],y=data1['time-to-60'],data=data1, hue = data1['brand'])
plt.title('cylinders vs time-to-60')
plt.xlabel('cylinders')
plt.ylabel('time-to-60')

plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.show()
```
